```python
#1)Numpy array manipulation
import numpy as np
#1.Create a 1d numpy array with integers from **1 to 24**.
array = np.arange(1, 25)
print(array)
```

```python
#2.Reshape the array into a 4x6 matrix.
array = array.reshape(4, 6)
print(array)
```

```python
#3.Extract the third row and second column from the reshaped array.
third_row = array[2, :]
second_column = array[:, 1]
print(third_row)
print(second_column)
```

```python
#4.Transpose the matrix and display the result.
transposed_array = array.T
print(transposed_array)
```

```python
#2)Array Element Extraction
#1.Create a 5x5 Numpy array with random integers between 10 and 100.
array = np.random.randint(10, 101, size=(5, 5))
print(array)
```

```python
#2.Extract all the even numbers from the array.
even_numbers = array[array % 2 == 0]
print(even_numbers)
```

```python
#3.Replace all numbers greater than 50 with 0.
array[array > 50] = 0
print(array)
```

```python
#4.Extract the diagonal elements of the array.
diagonal_elements = np.diag(array)
print(diagonal_elements)
```

```python
#3)DataFrame creation and operations(pandas)
#1.Create a pandas dataframe containing the following data.
# dictionary of lists
dict = {'Student_ID': [1,2,3,4,5],'Name':["Alice", "Bob", "Carol", "David","Eva"],
        'Subject': ["Math", "Science", "History", "Math","Science"],
        'Mark':[88, 75, 92, 85,90]}
import pandas as pd
df = pd.DataFrame(dict)
print(df.to_string(index=False))
```

```python
#2.Find the average marks for each subject.
average_marks = df.groupby('Subject')['Mark'].mean()
print(average_marks)
```

```python
#3.Add a new coloumn named Grade based on the marks:
import pandas as pd
data = {'Student_ID':['1','2','3','4','5'],'Name':["Alice", "Bob", "Carol", "David","Eva"],
        'Subject': ["Math", "Science", "History", "Math","Science"],
        'Mark':[88, 75, 92, 85,90]}
df = pd.DataFrame(data)
def calculate_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    else:
        return 'D'
df['Grade'] = df['Mark'].apply(calculate_grade)
df.index = pd.RangeIndex(start=1, stop=1+len(df))
print(df.to_string(index=False))


```

```python
#3.Find the mean and SD of the glucose levels.
import pandas as pd
df = pd.read_csv('/content/diabetes_model.csv')
glucose_mean = df['Glucose'].mean()
glucose_std = df['Glucose'].std()
print(f"Mean of Glucose Levels: {glucose_mean}")
print(f"Standard Deviation of Glucose Levels: {glucose_std}")
```

```python
#4.Plot a boxplot for the coloumn insulin.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/diabetes_model.csv')
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Insulin'])
plt.xlabel('Insulin')
plt.ylabel('Value')
plt.title('Boxplot of Insulin')
plt.show()

```

```python
#5.Create a heatmap showing the correlation between the numerical columns in the dataset.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/diabetes_model.csv')
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
